#### Step 1.1.1: Convert Missing Value Codes to NaNs
The fourth column of the feature attributes summary (loaded in above as `feat_info`) documents the codes from the data dictionary that indicate missing or unknown data. While the file encodes this as a list (e.g. `[-1,0]`), this will get read in as a string object. You'll need to do a little bit of parsing to make use of it to identify and clean the data. Convert data that matches a 'missing' or 'unknown' value code into a numpy NaN value. You might want to see how much data takes on a 'missing' or 'unknown' code, and how much data is naturally missing, as a point of interest.

**As one more reminder, you are encouraged to add additional cells to break up your analysis into manageable chunks.**

```python
# Build lookup for unknown values
expanded = []
for index, row in feat_info_c.iterrows():
    for v in row["missing_or_unknown"].replace('[','').replace(']','').split(','):
        expanded.append((row["attribute"],v))
    
moru_lkup = pd.DataFrame(expanded, columns=["attribute", "missing_or_unknown"])

moru_lkup.sort_values(["attribute", "missing_or_unknown"])
```

**Note:** Looking at the table split out by attribute name and unknown values, it can be easy to look up if you need. In this case, I did not use this, but it can be useful.

```python
# How many nans do we already have
natural_nans = []
df_length = len(azdias_c)
for c in azdias_c.columns:
    l = len(azdias_c[azdias_c[c].isna()][c])
    if l > 0:
        natural_nans.append([c, f'{l:,}', f'{df_length:,}', l/df_length])
    
df_nat_nan = pd.DataFrame(natural_nans, columns=["attribute", "nans", "total", "percentage"])

df_nat_nan["fPercentage"] = df_nat_nan.percentage.map('{:.2%}'.format)

df_nat_nan.sort_values(by="percentage", ascending=False) 
```

```python
# Identify missing or unknown data values and convert them to NaNs.
import json
import re

# Simple function to return NAN if this is in the nan feature info set
def get_na(i, j):
    if i in j:
        return np.nan
    else:
        return i

# Set unknowns to NaN 
for c in azdias_c.columns:
    try:
        feature = feat_info_c[feat_info_c.attribute==c].missing_or_unknown.values[0]
        sub = re.sub(r'\bX\b', '\"X\"', feature) # found these Xs in steps above, wasn;t enough to warrant general solution
        sub = re.sub(r'\bXX\b', '\"XX\"', sub) # found these XXs in steps above, wasn;t enough to warrant general solution
        j = json.loads(sub)
        azdias_c[c] = azdias_c[c].apply(lambda x: get_na(x, j))
    except:
        print("Failed to process column: {0} with values: {1}".format(c, sub))
    
print("Finished updating NaNs.")

```

### Step 1.3: Create a Cleaning Function

Even though you've finished cleaning up the general population demographics data, it's important to look ahead to the future and realize that you'll need to perform the same cleaning steps on the customer demographics data. In this substep, complete the function below to execute the main feature selection, encoding, and re-engineering steps you performed above. Then, when it comes to looking at the customer data in Step 3, you can just run this function on that DataFrame to get the trimmed dataset in a single step.

```python
def clean_data(df):
    """
    Perform feature trimming, re-encoding, and engineering for demographics
    data
    
    INPUT: Demographics DataFrame
    OUTPUT: Trimmed and cleaned demographics DataFrame
    """
    feat_info_copy = feat_info.copy()
    # Put in code here to execute all main cleaning steps:
    # convert missing value codes into NaNs, ...
    for c in df.columns:
        try:
            feature = feat_info_copy[feat_info_copy.attribute==c].missing_or_unknown.values[0]
            sub = re.sub(r'\bX\b', '\"X\"', feature) # found these Xs in steps above, wasn;t enough to warrant general solution
            sub = re.sub(r'\bXX\b', '\"XX\"', sub) # found these XXs in steps above, wasn;t enough to warrant general solution
            j = json.loads(sub)
            df[c] = df[c].apply(lambda x: get_na(x, j))
        except:
            print("Failed to process column: {0} with values: {1}".format(c, sub))
    
    # remove selected columns and rows, ...
    nans = []
    df_length = len(df[c])
    for c in df.columns:
        l = len(df[df[c].isna()][c])
        #if l > 0:
        nans.append([c, l, df_length, l/df_length])
    df_nan = pd.DataFrame(nans, columns=["attribute", "nans", "total", "percentage"])
    df_nan["fPercentage"] = df_nan.percentage.map('{:.2%}'.format)
    df_nan = df_nan.sort_values(by="percentage", ascending=False)
    anomolies = list(df_nan.head(6).attribute)
    df.drop(anomolies, axis=1, inplace=True)
    for c in anomolies:
        idx = feat_info_copy[feat_info_copy.attribute == c].index
        feat_info_copy.drop(index=idx, axis=0, inplace=True)    
    df["nbr_na_cols"] = df.isna().sum(axis=1)
    
    # select, re-encode, and engineer column values.
    feat_cats = feat_info_copy[feat_info_copy.type=='categorical'][['attribute', 'type']]
    vals = []
    for c in feat_cats.attribute:
        u = df[df[c].isna()==False][c].unique()
        vals.append([c, u, len(u), u.dtype])
    df_uniques = pd.DataFrame(vals, columns=['attribute', 'uniques', 'unique_count', 'data_type'])
    df_multi = df_uniques[~((df_uniques.unique_count==2)&(df_uniques.data_type!='object'))]
    df = pd.get_dummies(df,columns=df_multi.attribute)

    df['PRAEGENDE_JUGENDJAHRE_DEC'] = df.PRAEGENDE_JUGENDJAHRE.apply(lambda x: get_decade(x))
    df['PRAEGENDE_JUGENDJAHRE_MOV'] = df.PRAEGENDE_JUGENDJAHRE.apply(lambda x: get_movement(x))
    
    df.CAMEO_INTL_2015 = df.CAMEO_INTL_2015.astype(float) #convert to float for comparison
    df['CAMEO_INTL_2015_AGE'] = df.CAMEO_INTL_2015.apply(lambda x: get_age_group(x))
    df['CAMEO_INTL_2015_WEALTH'] = df.CAMEO_INTL_2015.apply(lambda x: get_wealth(x))
    
    # Return the cleaned dataframe.
    df.drop('PRAEGENDE_JUGENDJAHRE', axis=1, inplace=True)
    df.drop('CAMEO_INTL_2015', axis=1, inplace=True)
    df.drop('LP_LEBENSPHASE_FEIN', axis=1, inplace=True)
    df.drop('LP_LEBENSPHASE_GROB', axis=1, inplace=True)
    df.drop('nbr_na_cols', axis=1, inplace=True)
        
    del df_nan
    del df_length
    del feat_cats
    del df_uniques
    del df_multi
    return df
    
```

### Step 2.3: Interpret Principal Components

Now that we have our transformed principal components, it's a nice idea to check out the weight of each variable on the first few components to see if they can be interpreted in some fashion.

As a reminder, each principal component is a unit vector that points in the direction of highest variance (after accounting for the variance captured by earlier principal components). The further a weight is from zero, the more the principal component is in the direction of the corresponding feature. If two features have large weights of the same sign (both positive or both negative), then increases in one tend expect to be associated with increases in the other. To contrast, features with different signs can be expected to show a negative correlation: increases in one variable should result in a decrease in the other.

- To investigate the features, you should map each weight to their corresponding feature name, then sort the features according to weight. The most interesting features for each principal component, then, will be those at the beginning and end of the sorted list. Use the data dictionary document to help you understand these most prominent features, their relationships, and what a positive or negative value on the principal component might indicate.
- You should investigate and interpret feature associations from the first three principal components in this substep. To help facilitate this, you should write a function that you can call at any time to print the sorted list of feature weights, for the *i*-th principal component. This might come in handy in the next step of the project, when you interpret the tendencies of the discovered clusters.

```python
def plot_comps(pca, df, comp, defs):
    top_n = 5
    bot_n = 5
    weights = pca.components_
    cols = df.columns.values
    df_w1 = pd.DataFrame(list(zip(cols, weights[comp])), columns=['dim','weight'])
    df_w1['sort_col'] = abs(df_w1.weight)
    df_w1 = df_w1.sort_values('weight', ascending=False)[:top_n].append(
        df_w1.sort_values('weight', ascending=True)[:bot_n]).sort_values('weight', ascending=False)    
    if defs==None:
        df_w1['definitions'] =""
    else:
        df_w1['definitions'] = defs
        
    sns.barplot(x='weight',
                y="dim", data=df_w1, color='steelblue')
    plt.title("Component {0}".format(comp+1))
    plt.show()  
    return df_w1 
```
